# Анализ рынка видеоигр

Вы работаете в интернет-магазине, который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Необходимо выявить определяющие успешность игры закономерности на данных до 2016 года. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании на 2017-й год.


### Описание данных

- `Name	` — название игры
- `Platform	` — платформа
- `Year_of_Release	` — год выпуска
- `Genre	` — жанр игры
- `NA_sales	` — продажи в Северной Америке (миллионы проданных копий)
- `EU_sales	` — продажи в Европе (миллионы проданных копий)
- `JP_sales	` — продажи в Японии (миллионы проданных копий)
- `Other_sales	` — продажи в других странах (миллионы проданных копий)
- `Critic_Score	` — оценка критиков (максимум 100)
- `User_Score	` — оценка пользователей (максимум 10)
- `Rating	` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

```python
import pandas as pd
from scipy import stats as st
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

## Изучение общей информации.

```python
data = pd.read_csv('/datasets/games.csv')
display(data.info())
display(data.head(10))
display(data.isna().sum())
```

При первом взгляде видно некорректная запись названий столбцов, пропуски в половине столбцов и неподходящие для работы с данными форматы данных.

## Подготовка данных.

###  Замена названий столбцов.

```python
data.columns = data.columns.str.lower()
data.columns
```

### Преобразование типа данных.

```python
data.critic_score.unique()
```

Данные в столбцах `year_of_release` и `critic_score` можно заменить без других преобразований, поэтому замену для них объединим.

```python
col_change = ['year_of_release', 'critic_score']
for col in col_change:
    data[col] = data[col].astype('Int64')
    
```

```python
data.dtypes
```

```python
data.user_score.unique()
```

```python
data.user_score.value_counts()
```

В столбце `user_score`, как и в `critic_score`, должны находиться численные значения, но из-за строчки текстового значения 'tbd' изменен тип всего столбца. Гуглим несколько секунд и узнаем, что 'tbd' может расшифровываться как *'to betermined'* или *'to_be_decided'*. Если именно это означают 'tbd' в таблице, то рейтинга от пользователей ещё нет на момент составления таблицы, поэтому можем заменить их на NaNы, а затем весь столбец - на float.

```python
data.loc[data['user_score'] == 'tbd', 'user_score'] = np.nan
data.user_score = data.user_score.astype('float64')
data.info()
```

### Обработка пропусков.

```python
display(data.year_of_release.isna().sum())
display(data.genre.isna().sum())
```

Пропуски в столбце `year_of_release` составляют около процента от всего датасета. В столбце с жанрами их всего два, поэтому удаляем пропуски в обоих столбцах.

```python
data = data.dropna(axis='index', subset=['year_of_release', 'genre'])
data.info()
```

```python
data.rating.isna().sum()
```

Столбец `rating` - данные о рейтинге от негосударственной организации ESRB, расположенной в США и Канаде. Пропуски в данных обусловлены скорее всего тем, что оценки получили ограниченное количество игр из-за сложного процесса присвоения рейтинга.  Он инициируется издателем игры, после чего она долна пройти оценку от множества независимых рецензентов. Отталкиваясь от данных других столбцов, невозможно корректно подобрать данные для замены, потому заменим пропуски на `not_rated`

```python
data.rating = data.rating.fillna('not_rated')
data.rating.head(10)
```

```python
data.info()
```

В столбцах `critic_score` и `user_score` отсутствует слишком большое количество данных, которые нельзя заменить на основе имеющейся информации, но для дальнейших рассчетов их будет достаточно, поэтому пока что пропуски останутся. 

```python
#для проверки дубликатов приведем все данные в столбцах типа object к нижнему регистру:
columns=['name', 'platform', 'genre', 'rating']
for column in columns:
    data[column] = data[column].str.lower()
data.head()
```

```python
data.duplicated().sum()
```

Дубликаты могут прятаться в части датасета, поэтому попробуем их выявить по важным для нас столбцам:

```python
data[['name', 'platform', 'year_of_release']].duplicated().sum()
```

```python
data[['name', 'platform', 'year_of_release']] = data[['name', 'platform', 'year_of_release']].drop_duplicates()
```

```python
data[['name', 'platform', 'year_of_release']].duplicated().sum()
```

```python
data = data.dropna(axis='index', subset=['name', 'platform', 'year_of_release'])
```

```python
display(data.name.isna().sum())
display(data.platform.isna().sum()) 
display(data.year_of_release.isna().sum())
```

### Подсчет суммарных продаж во всех регионах и добавление нового столбца.

```python
data['total_sales'] = data[['na_sales','eu_sales','jp_sales', 'other_sales']].sum(axis=1)
data.head()
```

- Для подготовки данных к работе с ними названия столбцов перевели в нижний регистр методом .str.lower(). 
Заменили тип данных в столбцах `year_of_release`, `critic_score` на Int, более удобный формат для работы, чем float. А данные `user_score` из object - в float, после замены строк со значением "tbd" (to betermined) на NaNы методом .fillna().  
- Найдены пропуски в столбцах `year_of_release`, `genre`, `critic_score`, `user_score`, `rating`. Пропуски в первых двух можно просто удалить, так как составляют незначительную часть датасета. В `critic_score` и `user_score` пропуски оставим: пропущено много данных, и их нельзя заменить каким-либо корректным методом. 
- В столбце `rating` пропуски заменили на "not_rated", ведь пропуски в данном столбце означают отсутствие рейтинга негосударственной организации ESRB, расположенной в США и Канаде.
- Для вычисления неявных дубликатов привели все данные в столбцах типа object к нижнему регистру.

### Как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. 

Построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами для PS4, это популярная и прибыльная платформа.

```python
#создадим таблицу с данными только по ps4:
ps4 = top_5.loc[top_5['platform']=='ps4']
ps4.head()
```

```python

ps4.plot(kind='scatter', x='user_score', y='total_sales', figsize=(10,6));
plt.title('Зависимость продаж от оценки пользователей', color='blue')
plt.xlabel('Оценка игроков')
plt.ylabel('Продажи')
ps4.plot(kind='scatter', x='critic_score', y='total_sales', figsize=(10,6));
plt.title('Зависимость продаж от оценки критиков', color='blue')
plt.xlabel('Оценка критиков')
plt.ylabel('Продажи')
plt.show()
```

```python
#отделим только интересующие нас столбцы из таблицы ps4:
ps4_corr = ps4[['total_sales', 'user_score', 'critic_score']]
plt.figure(figsize=(1,3))
sns.heatmap(ps4_corr.corr()[['total_sales']].sort_values(by='total_sales', ascending=False), cmap="Greens", 
            annot=True, annot_kws={'size':15});
```

Связь роста продаж с высокими оценками критиков прослеживается лучше, чем с оценками пользователей. Матрица корреляции так же показывает слабую зависимость между продажами и оценками критиков и отсутствие какой-либо связи между отзывами пользователей и продажами. Это ожидаемый результат, так как пользователи могут оценить игру только после покупки, и очень популярная и ожидаемая игра может стать прибыльной для создателей, не понравившись при этом игрокам. 

Сравним выводы с продажами игр на других платформах.

```python
fig, axs = plt.subplots(5, 2, figsize=(20, 20))
plt.subplots_adjust(wspace=0.2, hspace=0.5)
x=0
for platform, games in top_5.groupby('platform'):
    
    top_5.plot(kind='scatter', x='critic_score', y='total_sales',
               title = 'Зависимость оценнок критиков и продаж, {}'.format(platform), fontsize=20, alpha=0.7, ax=axs[x, 0])
   
    top_5.plot(kind='scatter', x='user_score', y='total_sales',
               title = 'Зависимость оценок пользователей и продаж, {}'.format(platform), fontsize=17, alpha=0.7, ax=axs[x, 1])
    x+=1
        
```

```python
top_5_list = ['ps4', 'xone', '3ds', 'x360', 'ps3']
for name in top_5_list:
    platform = top_5.loc[top_5['platform']==name]
    platform_corr = platform[['total_sales', 'user_score', 'critic_score']]
    plt.figure(figsize=(1,3))
    ax = sns.heatmap(platform_corr.corr()[['total_sales']].sort_values(by='total_sales', ascending=False), cmap="Greens", 
            annot=True, annot_kws={'size':15});
    ax.set_title(name, fontsize=15)
```

Графики и коэффициенты корреляции по другим топовым платформам показывают схожий результат. Мнение пользователей так же не влияет на продажи, а оценка критиков слабо, но влияет на выручку.

### Распределение игр по жанрам.

- Какие жанры самые прибыльные?
- Выделяются ли жанры с высокими и низкими продажами?

```python
genre = data_period.groupby('genre').agg({'name': 'count', 'total_sales': 'sum'}).sort_values(by='total_sales', ascending=False).reset_index()
genre['total_sales'] = genre['total_sales']*10
genre.columns = ['genre', 'name', 'total_sales']

fig = plt.figure()
ax = fig.add_subplot(111)
ax2 = ax.twinx()
width = 0.7

ax.set_ylabel('Выпущенные игры, тыс. шт.')
ax=genre.name.plot(kind='bar',  figsize=(15,5), width=0.2, position=1, ec = 'black')
ax2=genre.total_sales.plot(kind='bar', figsize=(15,5),  color='yellow', width=0.2, position=0, ec = 'black')

ax2.set_ylabel('Суммарные продажи')
ax2.legend(['Количество игр', 'Сумма продаж'])
ax.set_title('Распределение продаж и количества игр по жанрам', fontsize=15)
plt.xticks(genre.index, genre.genre)
plt.grid(True)
plt.show()
```

По количеству проданных игр и сумме продаж лидирует жанр `action`. По продажам за ним идет жанр `shooter`, при чем количество выпущенных игр значительно меньше, чем в жанре `action`. Вероятно, спортивные игры стоят дороже в игр жанра `action`.

```python
def top_in_regions_plot(groupby, region_sales, ax):
    table = data_period.groupby(groupby).agg({region_sales: 'sum'}).sort_values(by=region_sales, ascending=False).head()
    title_dict = {'na_sales': 'Северная Америка', 'eu_sales': 'Европа', 'jp_sales': 'Япония'}
    plot = table.plot(kind='bar', ec='black', title=title_dict[region_sales], ax=axes[ax], 
                   fontsize=15)
    plot.legend(prop={'size':15})
    plot.set_xlabel('')
    plot.title.set_size(20)
```

## Проверка гипотез. 

**Гипотеза о различии средних пользовательских рейтингов Xbox One и PC.**

Сформулируем нулевую и альтернативную гипотезы:
 - H₀: средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
 - H₁: средние пользовательские рейтинги платформ Xbox One и PC различаются.

```python
xone_score = data_period[(data_period['platform'] == 'xone') & (data_period['user_score'] != 'NaN')]['user_score']
pc_score = data_period[(data_period['platform'] == 'pc') & (data_period['user_score'] != 'NaN')]['user_score']
#так как выборки независимы, будем использовать t-критерий Стьюдента
alpha = .05
results = st.ttest_ind(xone_score,
                      pc_score, equal_var = False)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

```python
display(round(xone_score.mean(), 2))
display(round(pc_score.mean(), 2))
```

Действительно, разница в средних оценках незначительна.

**Гипотеза о различии средних пользовательских рейтингов жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования»).**

Сформулируем нулевую и альтернативную гипотезы:
 - H₀: средние пользовательские рейтинги жанров Action и Sports равны;
 - H₁: средние пользовательские рейтинги жанров Action и Sports различаются.

```python
action_score = data_period[data_period['genre']=='action']['user_score'].dropna().reset_index(drop=True)
sports_score = data_period[data_period['genre']=='sports']['user_score'].dropna().reset_index(drop=True)
alpha = .05
results = st.ttest_ind(action_score,
                      sports_score, equal_var = False)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

```python
display(round(action_score.mean(), 2))
display(round(sports_score.mean(), 2))
```

P-значение намного меньше критического уровня статистической значимости, поэтому придерживаемся гипотезы о различии средних пользовательских оценах разных жанров. Простые вычисления подтверждают результаты проверки.
